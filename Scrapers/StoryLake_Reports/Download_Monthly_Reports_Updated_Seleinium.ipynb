```python
Download_dir = '/home/mbrenes/Downloads'
import os
import sys
import pandas as pd
from time import sleep
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import datetime as dt
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
def Download_Monthly(month,year):
    print("downloading report "+str(month)+"/"+year)
    Download_dir = '/home/mbrenes/Downloads'
    import os
    import sys
    import pandas as pd
    from time import sleep
    from selenium import webdriver
    from selenium.webdriver.support.ui import Select
    import datetime as dt

    browser = webdriver.Chrome()
    browser.get('https://owners.ciirus.com/Account/Login')
    sleep(5)
    # Login
    username= browser.find_element(by=By.XPATH, value="//input[@class='form-control'][@placeholder='Username']")
    password= browser.find_element(by=By.XPATH, value="//input[@class='form-control'][@placeholder='Password']")
    submit = browser.find_element(by=By.XPATH, value="//input[@type='submit']")
    username.send_keys("Mike1")
    password.send_keys("20Emily21@")
    submit.click()
    sleep(10)
    # Go to monthly report page and select month and year
    browser.get('https://owners.ciirus.com/Owner/OwnerMonthlyAccount')
    sleep(10)
    browser.get('https://owners.ciirus.com/Owner/OwnerMonthlyAccount')
    sleep(5)
    # get current window handle
    p = browser.current_window_handle
    # select month and year for report
    reportMonth =Select(browser.find_element(by=By.ID, value= "AccountMonth"))    
    reportYear =Select(browser.find_element(by=By.ID, value="AccountYear"))
    refresh =browser.find_element(by=By.ID, value="OwnerLoadMonthlyAccount")
    reportMonth.select_by_value(month)
    reportYear.select_by_value(year)
    refresh.click()
    sleep(4)
    # Download report 
    PrintReport =browser.find_element(by=By.ID, value="OwnerPrintMonthlyAccount")
    PrintReport.click()
    # takes a long time to load
    sleep(20)
    #Make print window active
    chwd = browser.window_handles
    for w in chwd:
        if w!=p:
            browser.switch_to.window(w)
    sleep(0.9)
    print("exporting report")
    export=browser.find_element(by=By.XPATH, value="//div[@title='Export To']")
    export.click()
    XLS=browser.find_element(by=By.XPATH, value="//div[@title='XLS']")
    XLS.click()
    sleep(5)
    # looks for all files in download folder created today. 
    newfile = 'Nothing '
    for file in os.listdir(Download_dir):
        if file == 'DxOwnerAccountMonthly.xls':
            oldfile = Download_dir+'/' + file
            newfile =month+'_'+year+'_Monthly_Report_Download.xls'
            os.rename(Download_dir+'/' + file, Download_dir+'/'+newfile )
            print(newfile +' downloaded')
```

```python
# PARSE OLD RENTAL ACTIVITY 
parsed= []
res = {}
for file in os.listdir('RentalActivity'):
    old_rr=pd.read_csv('RentalActivity/'+file)
    for i,rec in old_rr.iterrows():
        cleaner=rec['Description'].split(' ')[0]
        if cleaner == 'Rental':
            resID=rec['Description'].split('BID:')[1].split(')')[0].strip()                    
            arrival=rec['Description'].split(')')[1].split('-')[0].strip()
            departure = rec['Description'].split(')')[1].split('-')[1].strip()
            income=rec['Amount']
            res[resID] = {
            "arrival":arrival,
            "departure":departure,
            "income":income
            }
```
